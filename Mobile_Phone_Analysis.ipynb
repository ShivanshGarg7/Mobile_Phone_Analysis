```python
Top_10 = df.head(10)
```

```python
sns.set_theme()
plt.figure(figsize = (10,7))
graph = sns.barplot(data = Top_10, x = "units_sold_m", y = "model", hue = "smartphone",  palette=["Red", "Blue"])
```

```python
sns.set_theme()
plt.figure(figsize =(15,7))
sns.barplot(data = df.tail(10), x = "model", y = "units_sold_m")
```

```python
sns.set_theme()
plt.figure(figsize=(20, 12))
sns.countplot(data = df, y = "manufacturer", order = df["manufacturer"].value_counts().index)
```

## <span style = "color : Green"> **_In this Volume wise Plotting is done Samsung has the highest selling mobile phones market after that Nokia is in the 2nd Place._** </span>
***

```python
plt.figure(figsize = (15, 10))
sns.countplot(data = df, y = "form", order = df["form"].value_counts().index)
```

## <span style = 'color: green'> **_Here the most selling Mobile Phone type is Touchscreen and it has captured the Most of the Market after that Bar Mobile Phones is in the 2nd Place._** </span>
